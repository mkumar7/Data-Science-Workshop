```python
import pandas as pd
import datetime
import numpy as np
```

```python
df1 = pd.read_json("21FEB_ENTERED.json")
```

```python
df1
```

```python
df1.rename(columns = {'timestamp':'Entered_date_time'}, inplace = True) 
```

```python
df1
```

```python
df2 = pd.read_json("21FEB_INIT.json")
```

```python
df2
```

```python
df2.rename(columns = {'timestamp':'INIT_date_time'}, inplace = True) 
```

```python
df2
```

```python
df = pd.merge(df1, df2, on=["orderID"])
```

```python
df
```

```python
df['time_diff'] = df['Entered_date_time'] - df['INIT_date_time']
```

```python
df
```

```python
def my_func(x):
    y = str(x).split(":")
    _min = y[1]
    _sec = y[2].split("+")[0]
    return f'{_min}:{_sec}'
df['Order_time_in_Second'] = df['time_diff'].apply(my_func)
```

```python
df
```

```python
def year_month_day(date):
    return str(date).split(" ")[0]
df['order_date'] = df['INIT_date_time'].apply(year_month_day)
    
```

```python
df
```

```python
df['avg_order_time'] = pd.Series(pd.to_timedelta(df['time_diff'])).mean()
```

```python
df['avg_order_time']
```

```python
df.drop(['time_diff'], axis = 1)
```

```python
def my_func(x):
    y = str(x).split(":")
    _min = y[1]
    _sec = y[2].split("+")[0]
    return f'{_min}:{_sec}'
df['avg_time'] = df['avg_order_time'].apply(my_func)
```
